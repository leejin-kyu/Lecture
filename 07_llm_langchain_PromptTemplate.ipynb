```python
from langchain import PromptTemplate

# 특정 템플릿 문자열을 사용하여 PromptTemplate 인스턴스를 생성합니다
prompt = PromptTemplate(
    template="{product}는 어느 회사에서 개발한 제품인가요？",  # 'product'라는 자리 표시자를 포함하는 템플릿 문자열
    input_variables=[
        "product"  # 템플릿에서 사용될 변수 'product'를 정의합니다
    ]
)

# 'prompt' 객체를 사용하여 템플릿을 "소나타" 제품 이름으로 포맷하고 결과를 출력합니다
print(prompt.format(product="아반떼"))
print(prompt.format(product="소나타"))
print(prompt.format(product="벤츠"))
```
