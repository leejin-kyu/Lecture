```python
# konlpy 라이브러리에서 Okt 클래스를 가져옵니다. Okt는 텍스트의 형태소 분석을 도와줍니다.
from konlpy.tag import Okt

# collections 모듈에서 Counter 클래스를 가져옵니다. Counter는 리스트의 각 요소의 빈도수를 쉽게 카운트할 수 있습니다.
from collections import Counter

# wordcloud 라이브러리에서 WordCloud 클래스를 가져옵니다. WordCloud는 단어의 빈도수를 기반으로 워드 클라우드를 생성합니다.
from wordcloud import WordCloud

# matplotlib.pyplot을 plt로 가져옵니다. plt는 그래프나 이미지를 출력할 때 사용됩니다.
import matplotlib.pyplot as plt

# 분석할 텍스트를 지정합니다.
text = "나는 어제 학교에 갔다"

# Okt 클래스의 인스턴스를 생성합니다. 이를 통해 한글 텍스트의 형태소 분석을 수행합니다.
okt = Okt()

# 위에서 생성한 okt 인스턴스를 사용하여 텍스트에서 명사만 추출합니다.
nouns = okt.nouns(text)
print(nouns)  # 추출된 명사들을 출력합니다.

# Counter를 사용하여 추출된 명사들의 빈도수를 카운트합니다.
count = Counter(nouns)
print(count)  # 각 명사의 빈도수를 출력합니다.

# WordCloud를 사용하여 워드 클라우드를 생성합니다.
# font_path는 한글 폰트 경로를 지정해야 합니다. 여기서는 예시로 'C:/Windows/Fonts/malgun.ttf'라고 주석 처리했습니다.
wordcloud = WordCloud(font_path='/content/malgun.ttf',
                      background_color='white',
                      width=800,
                      height=600).generate_from_frequencies(count)

# matplotlib을 사용하여 워드 클라우드를 출력합니다.
plt.figure(figsize=(10, 8))          # 출력할 이미지의 크기를 지정합니다.
plt.imshow(wordcloud, interpolation='bilinear')  # 워드 클라우드 이미지를 출력합니다.
plt.axis('off')                      # 이미지의 축 정보를 표시하지 않도록 합니다.
plt.show()                           # 이미지를 화면에 보여줍니다.
```
